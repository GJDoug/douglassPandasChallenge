```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
totalPlayers = purchase_data["SN"].nunique()
totalPlayers
# print(f'Total Players = {totalPlayers}')

#create a dictionary for total players, build data frame and print
playersDict = {"Total Players" : [totalPlayers]}
playerFrame = pd.DataFrame(playersDict)
playerFrame
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#calculate measures and create dictionary
uniqueItems = purchase_data["Item Name"].nunique()
avgPrice = purchase_data["Price"].mean()
totalPurchases = purchase_data.loc[:,"SN"].count()
totalRevenue = purchase_data["Price"].sum()
summaryDict = {"# of Unique Items" : [uniqueItems],
                "Average Price" : [avgPrice],
                "Total # of Purchases" : [totalPurchases], 
                "Total Revenue" : [totalRevenue]
              }

#Format average price and total revenue
purchTotalFrame = pd.DataFrame(summaryDict)
purchTotalFrame["Average Price"] = purchTotalFrame["Average Price"].astype(float).map("${:,.2f}".format)
purchTotalFrame["Total Revenue"] = purchTotalFrame["Total Revenue"].astype(float).map("${:,.2f}".format)
purchTotalFrame
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#group genders, count players in each, and build data frame
genderGroup = purchase_data.groupby(['SN', 'Gender'])
uniquePlayers = pd.DataFrame(genderGroup.size())
uniquePlayers

#dataframe for unique players and total gender count
genderGroupCount = pd.DataFrame(uniquePlayers.groupby(['Gender']).count())
genderGroupCount.columns=['Total Count']


genderGroupCount["Percentage of Players"] = round(100*(genderGroupCount['Total Count']/genderGroupCount['Total Count'].sum()),1)
genderGroupCount

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
group = purchase_data.groupby('Gender')['SN'].nunique()
count = purchase_data.groupby('Gender')['Item ID'].count()
value = purchase_data.groupby('Gender')['Price'].sum()

purchAnlsys=count.to_frame('Purchase Count')
purchAnlsys['Average Purchase Price'] = (value/count).map('${:.2f}'.format)
purchAnlsys['Total Purchase Value'] = value.map('${:.2f}'.format)
purchAnlsys['Average total Purchase per Person'] = (value/group).map('${:.2f}'.format)
purchAnlsys
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
countSN = pd.DataFrame(purchase_data.groupby('SN').count())

sumSN = pd.DataFrame(purchase_data.groupby('SN').sum())
sumSN = sumSN.sort_values(by=['Price'], ascending=False)

sumSN['Purchase Count'] = countSN['Item ID']
sumSN['Average Purchase Price'] = round(sumSN['Price']/sumSN['Purchase Count'],2).map('${:,.2f}'.format)
sumSN['Total Purchase Value'] = sumSN['Price'].map('${:,.2f}'.format)

sumSN = sumSN.drop(labels = ['Purchase ID','Age','Item ID','Price'], axis=1)

sumSN.head(10)
```

```python
popularItems = purchase_data.filter(['Item ID', 'Item Name','Price'], axis=1)

avg_price = popularItems.groupby(['Item ID','Item Name'])['Price'].mean()
purchase_count = popularItems.groupby('Item ID')['Item Name'].value_counts()
purchase_value = avg_price * purchase_count

mostPopular = purchase_count.to_frame('Purchase Count')
mostPopular['Item Price'] = avg_price.map("${:.2f}".format)
mostPopular['Total Purchase Value'] = purchase_value

mostPopularA = mostPopular.sort_values(by="Purchase Count", ascending=True)
mostPopularA['Total Purchase Value'] = mostPopularA['Total Purchase Value'].map("${:.2f}".format)
mostPopularA
```
